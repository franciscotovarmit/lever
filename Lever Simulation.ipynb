# Lever - Fulcrum Gate simulation

## Modelling Fulcrum: A critical aspect to consider
The fulcrum plays a critical role and its modelling is not trivial.

We can simulate fulcrum's base as __fully fixed__ or through __contact elments__ (which allow interaction with neighbour surfaces and allow us to introduce friction parameters).

__a) Base fully fixed__ (As if it was "chemically" fixed to the floor). 

If we assume that the fulcrum is going to be fixed to the floor (by chemical or physical means such as chemical adhesion or plasma), we can simplify the model by restricting fulcrum's movement at its base. No need to add finite elements for the floor. Simpler and faster solution.

__b) Contact elements:__  Modelling fulcrum's - floor interaction.

We can also model it through contact algorithms.

When two or more bodies touch or collide, contact occurs between two surfaces of the bodies so that they cannot overlap in space. Numerically, this requires a special algorithm.  It allow us also to introduce surface properties such as friction parameters. There a few numerical parameters to adjust (tolerance of penetration, initial gap separation, contact algorithm, etc.). 

More info:  Kim NH. (2015) Finite Element Analysis for Contact Problems. In: Introduction to Nonlinear Finite Element Analysis. Springer, New York, NY. https://doi.org/10.1007/978-1-4419-1746-1_5


This is a much more refined and expensive solution (require more elements) but it gives a closer to reality results.  As a first approach, the floor has been modeled with PDMS mechanical properties.

Fulcrum's base is defined as "slave", while floor is defined as "Master". (It can be the opposite according to the problem and algorithm performance).
